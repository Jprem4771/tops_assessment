```python
import requests
from bs4 import BeautifulSoup
```

```python
web=requests.get("https://quotes.toscrape.com/")
```

```python
web
```

```python
soup = BeautifulSoup(web.content,"html.parser")
soup
```

```python
print(soup.prettify())
print(soup.title)
print(soup.a)
print(soup.p)
print(soup.h1)
print(soup.p.string)
```
